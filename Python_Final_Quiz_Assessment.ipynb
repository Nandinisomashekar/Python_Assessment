```python
import pandas as pd

# Creating a dataframe to store Quiz Questions and answers
quiz_questions = pd.DataFrame(columns = ["Question No.","Topic","Difficulty Level","Question","Options","Answer"])

# Creating dataframe to store users names and their scores
users = pd.DataFrame(columns = ["UserName","Scored Marks","Total Marks"])

# Function to add a question to the quiz_questions dataframe
# only super user is allowed to add a question
# Enter "super" when system asks the type of user
def add_question():
    print("\n")
    print("There are {0} questions".format(len(quiz_questions.index)))
    print("Enter the Questions in the order")
    print("Enter Question No")
    ques_no = input()
    print("Enter the Topic")
    topic= input()
    print("Enter Difficulty Level")
    diff_level= input()
    print("Enter the Question")
    ques= input()
    print("Enter the options")
    opt = input()
    print("Enter the Answer")
    ans = input()
    quiz_questions.loc[len(quiz_questions)] = [ques_no,topic,diff_level,ques,opt,ans]
    

# Function to attend the quiz
# Except super user all other users can attend the quiz
def attend_quiz():
    print("\n")
    print("There are {0} questions".format(len(quiz_questions.index)))
    print("Each question carries 10 marks")
    print("Total marks:{0}".format(len(quiz_questions.index)*10))
    print("\n")
    print("Enter UserName:")
    user_name = input()
    score=0
    report_correct =[]
    report_wrong = []
    for i in range(len(quiz_questions)):
        print("\n")
        print(quiz_questions.loc[i,"Question No."],quiz_questions.loc[i,"Question"])
        print(quiz_questions.loc[i,"Options"])
        print("Enter the Answer:")
        if input()==(quiz_questions.loc[i,"Answer"]):
            score+=1
            report_correct.append([quiz_questions.loc[i,"Question No."],quiz_questions.loc[i,"Question"],quiz_questions.loc[i,"Answer"]])
        else:
            report_wrong.append([quiz_questions.loc[i,"Question No."],quiz_questions.loc[i,"Question"],quiz_questions.loc[i,"Answer"]])
            pass
    users.loc[len(users)] = [user_name,score*10,len(quiz_questions.index)*10]
    print("\n")
    print("{0} your score is {1} out of {2}".format(user_name,score*10,len(quiz_questions.index)*10))
    print("\n")
    print("Questions which are Correctly Answered are")
    print("\n")
    if report_correct:
        for i in report_correct:
            print(i[0],i[1])
            print("   Correct Answer:{0}".format(i[2]))
    else:
        print("No Correct Answers")
    print("\n")
    print("Questions which are Wrongly Answered are")
    print("\n")
    if report_wrong:
        for j in report_wrong:
            print(j[0],j[1])
            print("   Correct Answer:{0}".format(j[2]))
    else:
        print("No Wrong Answers")
    
def check_superuser_or_user():
    print("Enter the user type:")
    if input()=="super":
        print("Super User")
        add_question()
    else:
        print("Normal User")
        attend_quiz()
        
check_superuser_or_user()
```
